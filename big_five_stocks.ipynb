```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
```

link to dataset on kaggle: https://www.kaggle.com/abdullahmu/big-five-stocks#big_five_stocks.csv

```python
# API token removed for privacy protection 

APPLE = requests.get('https://api.worldtradingdata.com/api/v1/history?symbol=AAPL&sort=newest&api_token=null')
GOOGLE = requests.get('https://api.worldtradingdata.com/api/v1/history?symbol=GOOGL&sort=newest&api_token=null')
MICROSOFT = requests.get('https://api.worldtradingdata.com/api/v1/history?symbol=MSFT&sort=newest&api_token=null')
AMAZON = requests.get('https://api.worldtradingdata.com/api/v1/history?symbol=AMZN&sort=newest&api_token=null')
FACEBOOK = requests.get('https://api.worldtradingdata.com/api/v1/history?symbol=FB&sort=newest&api_token=null')
```

```python
apple_df = pd.DataFrame(APPLE.json())
google_df = pd.DataFrame(GOOGLE.json())
microsoft_df = pd.DataFrame(MICROSOFT.json())
amazon_df = pd.DataFrame(AMAZON.json())
facebook_df = pd.DataFrame(FACEBOOK.json())
```

```python
facebook_df
```

```python
names_list=['Apple', 'Google', 'Microsoft', 'Amazon', 'Facebook']
```

```python
type(facebook_df['history'].iloc[0])
```

```python
history_keys = facebook_df['history'].iloc[0].keys()
history_keys
```

```python
for column in history_keys:
    apple_df[column] = apple_df[['history']].applymap(lambda x: float(x[column]))
# apple_df.drop(columns=['name', 'history'], inplace=True)
apple_df.drop(columns=['history'], inplace=True)

for column in history_keys:
    google_df[column] = google_df[['history']].applymap(lambda x: float(x[column]))
# google_df.drop(columns=['name', 'history'], inplace=True)
google_df.drop(columns=['history'], inplace=True)

for column in history_keys:
    microsoft_df[column] = microsoft_df[['history']].applymap(lambda x: float(x[column]))
# microsoft_df.drop(columns=['name', 'history'], inplace=True)
microsoft_df.drop(columns=['history'], inplace=True)

for column in history_keys:
    amazon_df[column] = amazon_df[['history']].applymap(lambda x: float(x[column]))
# amazon_df.drop(columns=['name', 'history'], inplace=True)
amazon_df.drop(columns=['history'], inplace=True)

for column in history_keys:
    facebook_df[column] = facebook_df[['history']].applymap(lambda x: float(x[column]))
# facebook_df.drop(columns=['name', 'history'], inplace=True)
facebook_df.drop(columns=['history'], inplace=True)
```

```python
apple_df.head()
```

```python
df = pd.concat([apple_df, google_df, microsoft_df, amazon_df, facebook_df], axis=0)
```
